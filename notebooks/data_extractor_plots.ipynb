```python
import sys
import re
import os
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
#total 
data = [['PCE', 482 + 37],
        ['VOC', 482 + 37],
        ['JSC', 482 + 37],
        ['FF', 482 + 37]]

total = pd.DataFrame(data=data, columns=['Property', 'Count'])
```

```python
#table_extractor + chemdataextractor
data = [['PCE', 430 + 28],
        ['VOC', 470 + 23],
        ['JSC', 460 + 21],
        ['FF', 465 + 24]]

tecde = pd.DataFrame(data=data, columns=['Property', 'Count'])
```

```python
#table_extractor 
data = [['PCE', 430],
        ['VOC', 470],
        ['JSC', 460],
        ['FF', 465]]

te = pd.DataFrame(data=data, columns=['Property', 'Count'])
```

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

```python
# load dataset
tips = sns.load_dataset("tips")

# # set plot style: grey grid in the background:
sns.set(style="whitegrid")

fig, ax = plt.subplots(1,1,figsize=(9, 9))

# # top bar -> sum all values(smoker=No and smoker=Yes) to find y position of the bars
# total = tips.groupby('day')['total_bill'].sum().reset_index()

# # bar chart 1 -> top bars (group of 'smoker=No')
bar1 = sns.barplot(x="Property",  y="Count", data=total, color='C1', ax=ax)
bar2 = sns.barplot(x="Property",  y="Count", data=tecde, color='C0', ax=ax)
bar3 = sns.barplot(x="Property",  y="Count", data=te, color='C2', ax=ax)

change_width(ax, 0.5)

# add legend
top_bar = mpatches.Patch(color='C1', label='Missing Data')
mid_bar = mpatches.Patch(color='C0', label='Extracted by ChemDataExtractor')
btm_bar = mpatches.Patch(color='C2', label='Extracted by Table Extractor')
plt.legend(handles=[top_bar, mid_bar, btm_bar], fontsize=20)

ax.set_ylim([0, 700])

ax.set_xlabel('Property', fontsize=20)
ax.set_ylabel('Counts', fontsize=20)
ax.tick_params(direction='out', length=6, width=2, labelsize=20)


fig.savefig('./all_extraction.png', bbox_inches='tight') 

# show the graph
plt.show()
```
